```python
import numpy as np
#[user][item]
a = np.array([[1, 4, 5, 0, 3],
             [5, 1, 0, 5, 2],
             [4, 1, 2, 5, 0],
             [0, 3, 4, 0, 4]])
a
```

```python
#Cosin thu cong:
users, items= a.shape
b = a.copy()
sim = []

for v1 in a:
  v = []
  for v2 in b:
       v.append(v1 @ v2 / (np.linalg.norm(v1) * np.linalg.norm(v2))) #Tinh goc giua 2 vector
  sim.append(v)


sim = np.array(sim)
sim
```

```python
#Cosin Thu Vien = Cosin Thu Cong
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(a, a)
```

```python
#Predict
k = 2
user = 4
item = 1
##########
```

```python
#Xep hang
user -= 1
item -= 1

user_predicted = [] #index - sim
for i in range(users):
  if a[i][item] != 0:
    user_predicted.append((i, sim[i][user]))

user_sorted = sorted(user_predicted, key=lambda x: x[1], reverse=True)
user_k = user_sorted[:k]
user_k
```

```python
def R(user): #Diem du doan trung binh cua tung user
  len = 0
  sum = 0
  for it in range(items):
    if a[user][it] != 0:
      sum += a[user][it]
      len += 1

  return sum/len
```

```python
#Person Sim
top = 0
bot = 0
for i,sim in user_k:
  top +=  sim*(a[i][item] - R(i))
  bot += sim

predict = R(user) + top / bot
predict
R(user)
```

## <1> UserBase

```python
#@title Input
a = np.array([[1, 4, 5, 0, 3],
             [5, 1, 0, 5, 2], #hang 2 va hang 3 tuong quan
             [4, 1, 2, 0, 0], #bo di so 5 de check xem du doan co gan voi 5 khong
             [0, 3, 4, 0, 4]])
#Predict
k = 2 #@param {type:""}
user = 3 #@param {type:""}
item = 4 #@param {type:""}
##########
a
```

```python
#@title Process
def R(user): #Diem du doan trung binh cua tung user
  len = 0
  sum = 0
  for it in range(items):
    if a[user][it] != 0:
      sum += a[user][it]
      len += 1

  return sum/len

def user_base(a, k, user, item):

  # Cosin thu cong:
  users, items= a.shape
  b = a.copy()
  sim = []

  for v1 in a:
    v = []
    for v2 in b:
        v.append(v1 @ v2 / (np.linalg.norm(v1) * np.linalg.norm(v2))) #Tinh goc giua 2 vector
    sim.append(v)
  sim = np.array(sim)

  # Cosin Thu Vien
  # from sklearn.metrics.pairwise import cosine_similarity
  # cosine_similarity(a, a)

  #Xep hang
  user -= 1
  item -= 1

  user_predicted = [] #index, sim
  for i in range(users):
    if a[i][item] != 0:
      user_predicted.append((i, sim[i][user]))

  user_sorted = sorted(user_predicted, key=lambda x: x[1], reverse=True)
  user_k = user_sorted[:k]

  #Person Sim
  top = 0
  bot = 0
  for i,sim in user_k:
    top +=  sim*(a[i][item] - R(i))
    bot += sim

  predict = R(user) + top / bot
  return predict
```

```python
#@title Output
user_base(a, k, user, item) #Ket qua gan voi so 5 da loai bo
```

## <3> Baseline Predictor



```python
#@title Input
x = np.nan
a = np.array([[1, 4, 5, x, 3],
             [5, 1, x, 5, 2], #hang 2 va hang 3 tuong quan
             [4, 1, 2, x, x], #bo di so 5 de check xem du doan co gan voi 5 khong
             [x, 3, 4, x, 4]])
#Predict
user = 3 #@param {type:""}
item = 4 #@param {type:""}
##########
a
```

```python
#@title Process
def baseline(a, user, item):

  _u = np.nanmean(a)
  bias_user = np.nanmean(a[user-1, :]) - _u
  bias_item = np.nanmean(a[:, item-1]) - _u

  return _u + bias_user + bias_item
```

```python
#@title Output
baseline(a, user, item) #Ket qua gan voi so 5 da loai bo
```

##<4> Matrix factorization

```python
import numpy as np
a =  np.array([[1, 4, 5, 0, 3],
              [5, 1, 0, 5, 2], #hang 2 va hang 3 tuong quan
              [4, 1, 2, 0, 0], #bo di so 5 de check xem du doan co gan voi 5 khong
              [0, 3, 4, 0, 4]])


#latent_factors
K = 2  #@param {type:""}

#leaning_rate
beta = 0.01 #@param {type:""}

#regularization
lamda = 0.02 #@param {type:""}

epos = 1000 #@param {type:""}
a
```

```python
def matrix_factorization(a, K, beta, lamda, epos):
  users, items = a.shape

  W = np.random.rand(users, K)
  H = np.random.rand(items, K)


  # Training
  for step in range(epos):
      for u in range(users):
          for i in range(items):
              if a[u, i] > 0:
                  error = a[u, i] - np.dot(W[u, :], H[i, :])
                  for k in range(K):
                      W[u, k] += beta * (error * H[i, k] - lamda * W[u, k])
                      H[i, k] += beta * (error * W[u, k] - lamda * H[i, k])
      print('epos',step,'loss:',error)
  return np.dot(W, H.T)
```

```python
matrix_factorization(a, K, beta, lamda, epos) #Ket qua a[3][4] gan voi so 5 da loai bo
```

## <5> Matrix factorization Upgare

```python
import numpy as np
x = np.nan
a =  np.array([[1, 4, 5, x, 3],
              [5, 1, x, 5, 2], #hang 2 va hang 3 tuong quan
              [4, 1, 2, x, x], #bo di so 5 de check xem du doan co gan voi 5 khong
              [x, 3, 4, x, 4]])


#latent_factors
K = 2  #@param {type:""}

#leaning_rate
beta = 0.01 #@param {type:""}

#regularization
lamda = 0.02 #@param {type:""}

epos = 1000 #@param {type:""}
a
```

```python
def matrix_factorization_upgare(a, K, beta, lamda, epos):

  users, items = a.shape

  W = np.random.rand(users, K)
  H = np.random.rand(items, K)

  #upgare
  _u = np.nanmean(a)

  bias_users = []
  bias_items = []
  for u in range(users):
    bias_users.append(np.nanmean(a[u, :]) - _u)

  for i in range(items):
    bias_items.append(np.nanmean(a[:, i]) - _u)
  #######

  a[np.isnan(a)] = 0
  print(a)
  # Training
  for step in range(epos):
      for u in range(users):
          for i in range(items):
              if a[u, i] > 0:

                  aui = _u + bias_users[u] + bias_items[i] + np.dot(W[u, :], H[i, :])

                  error = a[u, i] - aui
                  _u = _u + beta * error
                  bias_users[u] = bias_users[u] + beta * (error - lamda * bias_users[u])
                  bias_items[i] = bias_items[i] + beta * (error - lamda * bias_items[i])

                  for k in range(K):

                      W[u, k] += beta * (error * H[i, k] - lamda * W[u, k])
                      H[i, k] += beta * (error * W[u, k] - lamda * H[i, k])
      print('epos',step,'loss:',error)
  matrix = np.dot(W, H.T)
  for u in range(users):
    for i in range(items):
      matrix[u][i] += _u + bias_users[u] + bias_items[i]
  return matrix
```

```python
matrix_factorization_upgare(a, K, beta, lamda, epos) #Ket qua a[3][4] gan voi so 5 da loai bo
```
